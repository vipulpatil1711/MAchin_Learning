```python
import pandas as pd
import numpy as np
from sklearn. preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
df = pd.read_csv('D:\Python\datasets\Hotel Reservations.csv')
df.head(5)
```

```python
df.columns
```

```python
df.drop(columns=['booking_status'],inplace=True)
```

```python
df.head(5)
```

```python
df.info()
```

```python
df_feature = df
df_feature.head(3)
```

```python
numeric_features=['no_of_adults','no_of_children','no_of_weekend_nights','no_of_week_nights','required_car_parking_space',
                  'lead_time','arrival_year','arrival_month','arrival_date','repeated_guest','no_of_previous_cancellations',
                  'no_of_previous_bookings_not_canceled','avg_price_per_room','no_of_special_requests']

Cat_features = ['Booking_ID','type_of_meal_plan','room_type_reserved','market_segment_type']                   
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline

preprocessor = ColumnTransformer(transformers=[("num", StandardScaler(), numeric_features),
        ("cat", OneHotEncoder(handle_unknown="ignore"), Cat_features),])
```

```python
dbscan = DBSCAN( eps =1.5,min_samples=5,metric="euclidean")
```

```python
model = Pipeline(steps=[("preprocess",preprocessor),
                         ("cluster",dbscan)])
```

```python
model.fit(df_feature)
```

```python
labels = model.named_steps['cluster'].labels_
df['dbscan_cluster'] = labels
```
